```python
#Create the zone of the shot column and intialized all the column values to zero
shotdf['shotzone'] = 0
```

```python
#1 stands for corner 3
#2 stands for non corner 3's
#3 stands for two point shots 

def defZone(i, x, y):
    #corner threes 
    if y <= 7.8 and ((x > 22) or (x < -22)):
        shotdf['shotzone'][i] = "Corner 3"
    elif y > 7.8 and (x*x + y*y) > 23.75*23.75:
        shotdf['shotzone'][i] = "Non Corner 3"
    else:
        shotdf['shotzone'][i] = "Two Point"
```

```python
for i in range(len(shotdf)):
    defZone(i, shotdf.iloc[i,1], shotdf.iloc[i,2])
```

3. Given the defZone() we can caluculate the shot distribution 

```python
shotdf.groupby(["team"])["fgmade"].count()
```

```python
shotdf.groupby(["team", "shotzone"])["fgmade"].count()
```

```python
#shot distributution
shot_dis = (shotdf.groupby(["team", "shotzone"])["fgmade"].count()/shotdf.groupby(["team"])["fgmade"].count()).to_frame()
shot_dis.reset_index(inplace=True)
shot_dis.rename(columns = {'fgmade':'Shot Distribution'}, inplace = True)
shot_dis
```

4. Calculate shot percentage and eFG% for each zone

```python
dfPercentage = (shotdf.groupby(["team", "shotzone"])["fgmade"].sum()/shotdf.groupby(["team", "shotzone"])["fgmade"].count()).to_frame()
dfPercentage.reset_index(inplace=True)
dfPercentage
```

```python
dfEFG = (shotdf.groupby(["team", "shotzone"])["fgmade"].sum()*1.5/shotdf.groupby(["team", "shotzone"])["fgmade"].count()).to_frame()
dfEFG.reset_index(inplace=True)
dfEFG
```

```python
dfEFG.loc[2,'fgmade'] = dfPercentage.loc[2,'fgmade']
dfEFG.loc[5,'fgmade'] = dfPercentage.loc[5,'fgmade']
```

```python
dfPercentage.rename(columns = {'fgmade':'Shot Percentage'}, inplace = True)
dfPercentage
```

```python
dfEFG.rename(columns = {'fgmade':'eFG%'}, inplace = True)
dfEFG
```
